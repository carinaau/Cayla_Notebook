```python
import pandas as pd
import numpy as np
df = pd.read_csv('Diabetes.csv')
df.info()
```

```python
df.head()
```

```python
df = df.fillna(0)
```

```python
df2 = df[['Glucose','BMI','Age','Outcome']]
```

```python
df2.head()
```

```python
df2.describe()
```

```python
df2[df2.columns[:-1]]==0
#取前三列的数据
```

```python
df3 = df2.loc[~(df2[df2.columns[:-1]]==0).any(axis=1)]
df3.describe()

#df2[df2.columns[:-1]]==0：对 DataFrame df2 中除了最后一列之外的所有列进行逐元素的零值检查
#.any(axis=1)：对上一步得到的布尔型 DataFrame 沿着行的方向（axis=1）进行 any 操作，即只要该行中存在任意一个 True，就返回 True。
#~( ... )：使用波浪号 ~ 对上一步结果取反，得到的是一个布尔型 Series，表示哪些行中所有列的值都不等于零。
#df2.loc[...]：通过布尔索引选择符合条件的行，从原始 DataFrame df2 中提取出符合条件的行，存储在新的 DataFrame df3 中。
```

```python
df3.info()
#比df减少了十几行，数据质量upupup
```

```python
df3.groupby('Outcome').agg({'Glucose':'mean','BMI':'median','Age':'mean'})
```
